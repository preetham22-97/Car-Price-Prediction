```python
final_dataset= df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_dataset.columns
```

```python
final_dataset.head()
```

```python
final_dataset['Current Year']=2020
```

```python
final_dataset.head()
```

```python
final_dataset['No of years'] = final_dataset['Current Year'] - final_dataset['Year']
```

```python
final_dataset.head()
```

```python
final_dataset.drop(['Year'],axis=1,inplace =True)
```

```python
final_dataset.drop(['Current Year'],axis=1,inplace =True)
```

```python
final_dataset.head()
```

```python
## Creating dummy variable for categorical features
final_dataset= pd.get_dummies(final_dataset,drop_first=True)
```

```python
final_dataset.head()
```

```python
## Create Independent and Dependent Features
X = final_dataset.iloc[:,1:]
y = final_dataset.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
## Features Importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
feature_importances = pd.Series(model.feature_importances_,index=X.columns)
feature_importances.nlargest(5).plot(kind='barh')
plt.xlabel('Correlation Value')
plt.title('Top 5 Important features')
```

```python
## Create train and test data
from sklearn.model_selection import train_test_split
X_train , X_test , y_train ,y_test = train_test_split(X,y,test_size=0.3,random_state=420)
```

```python
print('X_train  : {}'.format(X_train.shape))
print('X_test  : {}'.format(y_train.shape))
print('y_train  : {}'.format(X_test.shape))
print('y_test  : {}'.format(y_test.shape))
```

```python
## Implementing RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=20, min_samples_split=15, n_estimators=700)
```

```python
## Hyperparameter Tuning
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
print(n_estimators)
```

```python
##List of Parameters for Hyparameter Tuning

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
## Create a dictionary for Parameters to find the best parameters
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
## Importing library for RandomizedSearchCV with 5-CrossValidation
from sklearn.model_selection import RandomizedSearchCV
rf_randomcv = RandomizedSearchCV(estimator =rf , param_distributions = random_grid,scoring='neg_mean_squared_error',cv=5,verbose=5,random_state=42)
```

```python
rf_randomcv.fit(X_train,y_train)
```

```python
## Checking out the best values for the parameters
print(rf_randomcv.best_estimator_)
print(rf_randomcv.best_params_)
```

```python
## Predicting the selling price
rf.fit(X_train,y_train)
predictions = rf.predict(X_test)
```

```python
predictions
```

```python
sns.distplot(y_test - predictions)
```

```python
sns.set_style('whitegrid')

sns.regplot(y_test,predictions,fit_reg=True,color='red',marker='D')
```
