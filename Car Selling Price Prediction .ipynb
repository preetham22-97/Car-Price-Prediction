### Random Forest Regression

```python
#Finding the importance of each feature given in the dataset
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_independent,y_dependent)
```

```python
#Used feature_importance attribute
print(model.feature_importances_)
```

```python
feature_importances = pd.Series(model.feature_importances_,index=X_independent.columns)
feature_importances.nlargest(5).plot(kind='bar', color='Green')
plt.xlabel('Correlation Value')
plt.title('Top five most Important Features')
```

```python
#Create Both Training Dataset and Test Data set in 70:30 ratio
from sklearn.model_selection import train_test_split
X_train , X_test , y_train ,y_test = train_test_split(X_independent,y_dependent,test_size=0.3,random_state=420)
```

```python
print('X_train  : {}'.format(X_train.shape))
print('X_test  : {}'.format(y_train.shape))
print('y_train  : {}'.format(X_test.shape))
print('y_test  : {}'.format(y_test.shape))
```

```python
## Implementing RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=20, min_samples_split=15, n_estimators=700)
```

```python
#Implementing Hyperparameter Tuning to find the best parameters for the model. Helps big time in obtaining high accuracy
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
print(n_estimators)
```

```python
#List of Parameters for Hyparameter Tuning

#Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
#Number of features to consider at every split
max_features = ['auto', 'sqrt']
#Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

#Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
#Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
#Create a dictionary for Parameters to find the best parameters
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
#Using RandomizedSearchCV with 5-CrossValidation technique
from sklearn.model_selection import RandomizedSearchCV
rf_randomcv = RandomizedSearchCV(estimator =rf , param_distributions = random_grid,scoring='neg_mean_squared_error',cv=5,verbose=5,random_state=42)
```

```python
rf_randomcv.fit(X_train,y_train)
```

```python
## Checking out the best values for the parameters
print(rf_randomcv.best_estimator_)
print(rf_randomcv.best_params_)
```

```python
#Predicting Selling price using Test Data
rf.fit(X_train,y_train)
predictions = rf.predict(X_test)
```

```python
predictions
```

```python
sns.histplot(y_test - predictions,bins=10)
```

```python
sns.set_style('dark')

sns.regplot(y_test,predictions,fit_reg=True,color='blue',marker='s')
```
